```python
import pandas as pd
from pyecharts.charts import Geo
from pyecharts.charts import Map
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType,ThemeType
```

```python
df = pd.read_csv(r"Distribution of residence.csv") 
df
```

```python
城市富豪居住地人数 = list(zip(list(df.居住地),list(df.人数)))
城市富豪居住地人数
```

```python
keys = list(df.居住地)
keys
```

```python
values = list(df.人数)
values
```

```python
def map_visualmap() -> Map:
    c = (
        Map()
        .add("人数",城市富豪居住地人数, "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="城市富豪居住地人数"),
            visualmap_opts=opts.VisualMapOpts(max_=50),
        )
    )
    return c
map_visualmap().render_notebook()
```
